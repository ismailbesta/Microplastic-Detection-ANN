```python
import zipfile
import os
from pathlib import Path

# Lightning AI'da Google Drive olmadığı için zip dosyanın 
# notebook ile aynı klasörde olduğunu varsayıyoruz:
ZIP_PATH    = 'merged_dataset.zip' 
DATASET_DIR = 'dataset' # Baştaki /content/ ibaresini kaldırdık

# Klasörü oluştur ve zipten çıkar
os.makedirs(DATASET_DIR, exist_ok=True)
with zipfile.ZipFile(ZIP_PATH, 'r') as z:
    z.extractall(DATASET_DIR)

print("Zipten çıkarma işlemi tamamlandı!")

# Klasör yapısını doğrulama (YOLO formatı kontrolü)
for split in ['train', 'valid', 'test']:
    # Not: Zipten çıkınca içinden doğrudan 'merged_dataset' klasörü mü çıkıyor, 
    # yoksa direkt resimler mi geliyor? Eğer iç içe klasör oluştuysa yol doğrudur.
    # Eğer direkt çıktıysa 'merged_dataset/' kısmını yoldan silebilirsin.
    img_path = Path(f'{DATASET_DIR}/merged_dataset/{split}/images')
    label_path = Path(f'{DATASET_DIR}/merged_dataset/{split}/labels')
    
    if img_path.exists() and label_path.exists():
        imgs = len(list(img_path.glob('*')))
        labels = len(list(label_path.glob('*')))
        print(f'{split}: {imgs} görüntü, {labels} label')
    else:
        print(f'Hata: {split} dizini bulunamadı. Lütfen zipten çıkan klasör yapısını kontrol et.')
```

```python
import yaml
import os

# DATASET_DIR'in tanımlı olduğundan emin ol (Örn: 'dataset')
DATASET_DIR = 'dataset' 

# data.yaml dosyasının tam (absolute) yolunu alıyoruz
yaml_path = os.path.abspath(f'{DATASET_DIR}/merged_dataset/data.yaml')

# YAML dosyasını oku
with open(yaml_path, 'r', encoding='utf-8') as f:
    cfg = yaml.safe_load(f)

# Klasörlerin sistemdeki tam/mutlak yollarını (absolute path) çıkarıyoruz
# Lightning AI bu sayede klasörleri eğitme esnasında hatasız bulur.
cfg['train'] = os.path.abspath(f'{DATASET_DIR}/merged_dataset/train/images')
cfg['val']   = os.path.abspath(f'{DATASET_DIR}/merged_dataset/valid/images')
cfg['test']  = os.path.abspath(f'{DATASET_DIR}/merged_dataset/test/images')

# Eğer yaml içinde eski 'path' anahtarı kalmışsa çakışma yaratmaması için siliyoruz
if 'path' in cfg:
    del cfg['path']

# Güncellenmiş YAML dosyasını kaydet
with open(yaml_path, 'w', encoding='utf-8') as f:
    yaml.dump(cfg, f, default_flow_style=False, allow_unicode=True)

print("data.yaml başarıyla güncellendi:")
print(open(yaml_path, 'r', encoding='utf-8').read())
```

```python
from ultralytics import YOLO
import os

# Model yükleme
model = YOLO('yolov8s.pt')  # small model

# Çıktıların (weights, plots vb.) kaydedileceği klasörü Lightning AI'a göre ayarlıyoruz
# Bu klasör sol taraftaki dosya gezgininde otomatik oluşacaktır.
project_dir = os.path.abspath('mikroplastik_egitimleri')

results = model.train(
    # ── Veri ──────────────────────────────────────────
    data      = yaml_path,    # Bir önceki adımda güncellediğimiz mutlak yol
    imgsz     = 1280,
    batch     = 16,
    workers   = 4,            # Lightning AI sunucuları güçlüdür, 2'den 4'e çıkararak hızlandırabilirsin

    # ── Eğitim süresi ─────────────────────────────────
    epochs    = 50,           # Yorum satırındaki gibi 100 de yapabilirsin, early stopping (patience) korur
    patience  = 15,           

    # ── Optimizer ─────────────────────────────────────
    optimizer = 'AdamW',      
    lr0       = 0.001,        
    lrf       = 0.01,         
    momentum  = 0.937,        
    weight_decay = 0.0005,    

    # ── Cosine LR Scheduler ───────────────────────────
    cos_lr    = True,         

    # ── Augmentation ──────────────────────────────────
    mosaic    = 1.0,          
    mixup     = 0.15,         
    copy_paste= 0.1,          
    degrees   = 15.0,         
    translate = 0.1,          
    scale     = 0.5,          
    shear     = 2.0,          
    perspective = 0.0001,     
    flipud    = 0.5,          
    fliplr    = 0.5,          
    hsv_h     = 0.015,        
    hsv_s     = 0.7,          
    hsv_v     = 0.4,          

    # ── Regularizasyon ────────────────────────────────
    dropout   = 0.1,          

    # ── Kayıt (LIGHTNING AI UYARLAMASI) ───────────────
    project   = project_dir,   # Google Drive yerine yerel proje klasörü
    name      = 'yolov8s_adamw',
    save      = True,
    save_period = 10,         
    plots     = True,         

    # ── Küçük nesne desteği ───────────────────────────
    overlap_mask = True,
    verbose   = True,
    
    # ── Donanım ───────────────────────────────────────
    device    = 0             # Eğer makinede GPU varsa direkt ilk GPU'yu (cuda:0) kullanır
)

# Metrikleri yazdırma (Sözlük anahtarlarında Ultralytics sürümlerine göre küçük farklar olabiliyor, güvenli almak için koruduk)
try:
    print(f"\nmAP@50:    {results.results_dict.get('metrics/mAP50(B)', 'N/A'):.4f}")
    print(f"mAP@50-95: {results.results_dict.get('metrics/mAP50-95(B)', 'N/A'):.4f}")
except Exception:
    # Eğer yukarıdaki anahtarlar boş dönerse genel sonuç nesnesinden yazdırır
    print("\nEğitim başarıyla tamamlandı. Metrikler kaydedildi.")
```
